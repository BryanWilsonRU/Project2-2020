```python
# Dependencie
import pandas as pd
import numpy as np
```

```python
# Load in file
poverty_file = "PeopleBelowPovertyLevel.csv"
# Read and display the CSV with Pandas
poverty_file_df = pd.read_csv(poverty_file)
poverty_file_df.head()
```

```python
poverty_file_df
```

```python
#Strip blank spaces using Lambda funcion
poverty_file_df["Geographic Area"]=poverty_file_df["Geographic Area"].apply(lambda x: x.strip())
poverty_file_df["City"]=poverty_file_df["City"].apply(lambda x: x.strip())
poverty_file_df["poverty_rate"]=poverty_file_df["poverty_rate"].apply(lambda x: x.strip())
#Display results-final df
poverty_file_df.head()
```

```python
S1= poverty_file_df.drop(columns ='City')
```

```python
#Drop all rows
S2 = S1.dropna(how='any')
S2.head()
```

```python
#List of columns to create db
S2.dtypes
```

```python
S2.count()
```

```python
S3 = S2[~S2.poverty_rate.str.contains("-")]
S3.count()
```

```python
#Rename the column name
S4 = S3.rename(columns={"Geographic Area" : "State"})
S4.head()
```

```python
#List of columns to create db
S4.dtypes
```

```python
S4[["poverty_rate"]] = S4[["poverty_rate"]].apply(pd.to_numeric)
S4.dtypes
```

```python
S5 = S4.groupby('State').poverty_rate.mean()
S5.head()
```

```python
cleaned_Poverty_file= S5.to_frame()
cleaned_Poverty_file.head()
```

```python
cleaning_Done = cleaned_Poverty_file.rename(columns={"poverty_rate" : "poverty_rate in (%)"})
cleaning_Done
```

```python
cleaning_Done.to_csv(r'/Users/hasti/Desktop/Poverty_rate.csv')
```
