```python
# Load the Pandas libraries with alias 'pd' 
import pandas as pd 
# Read data from file 'filename.csv' 
PoliceDeaths = pd.read_csv("police_deaths_538UTF.csv") 
# Preview the first 5 lines of the loaded data 
PoliceDeaths.head()
```

```python
Step1 = PoliceDeaths.drop(['person', 'dept', 'eow', 'cause', 'date', 'dept_name'], axis = 1)
Step1.head()
```

```python
Step2 = Step1[Step1.canine != True]
Step2
```

```python
Step3 = Step2[Step2['year'] > 2014]
Step3.count()
```

```python
Step4 = Step3.drop(['canine'], axis = 1)
Step4.head()
```

```python
CleanPoliceDeaths = Step4.rename(columns={"state": "State"})
CleanPoliceDeaths.head()
```

```python
CleanPoliceDeaths = CleanPoliceDeaths.iloc[3:]
CleanPoliceDeaths.set_index('State', inplace = True)
CleanPoliceDeaths.head()
```
