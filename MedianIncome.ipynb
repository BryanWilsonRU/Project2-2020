```python
import pandas as pd
import numpy as np
```

```python
mhi_df = pd.read_csv("Project2_Data/datasets_2647_4395_MedianHouseholdIncome2015.csv", engine='python')
mhi_df
```

```python
mhi_df.count()
```

```python
mhi_df = mhi_df.dropna(how='any')
mhi_df.head(60)
```

```python
mhi_df.count()
```

```python
mhi_df.dtypes
```

```python
mhi_df = mhi_df.drop(mhi_df.index[[44,0]])
# mhi_df = mhi_df.drop(mhi_df.index[[57,0]])
mhi_df.head(60)
```

```python
x = mhi_df.loc[mhi_df["Median Income"] == "(X)", :]
x
```

```python
mhi_df = pd.concat([mhi_df, x, x]).drop_duplicates(keep=False)
mhi_df
```

```python
dash = mhi_df.loc[mhi_df["Median Income"] == "-", :]
dash
```

```python
mhi_df = pd.concat([mhi_df, dash, dash]).drop_duplicates(keep=False)
mhi_df.head(60)
```

```python
dumb = mhi_df.loc[mhi_df["Median Income"] == "2,500-",:]
```

```python
mhi_df = pd.concat([mhi_df, dumb, dumb]).drop_duplicates(keep=False)
```

```python
mhi_df = mhi_df.rename(columns = {"Geographic Area": "State",
                                  "Median Income": "Median_Income"})
```

```python
dumber = mhi_df.loc[mhi_df["Median_Income"] == "250,000+",:]
mhi_df = pd.concat([mhi_df, dumber, dumber]).drop_duplicates(keep=False)
```

```python
mhi_df["Median_Income"] = pd.to_numeric(mhi_df["Median_Income"])
```

```python
mhi_df.dtypes
```

```python
mhi_df.to_csv("Clean_Data/median_household_income.csv", header=True)
```
